```python
# *** Functions ***
# Declare a function 'Compute' that takes two arguments:
# distance and time, and use it to calculate velocity.

def Compute(distance, time):
    velocity = distance / time
    return velocity

# Here:  Distance = 250 m  and  Time = 20 s
print("The velocity is", Compute(250, 20), "m/s")
```

```python
# *** Forloop ***
# Declare a list 'even' that contains all even numbers up till 16.

even = [0,2,4,6,8,10,12,14,16]

# Declare a function 'sum' that takes the list as an argument and
# Calculates the sum of all entries using a for loop.

def sum(even):
    total = 0
    for num in even:
        total += num
    return total

sum(even)
```

```python
#  *** Plotting a Single Line: ***
# Compute the x and y coordinates for points on a sine curve and
# Plot the points using matplotlib

import numpy as np
import matplotlib.pyplot as plt

x = np.arange(0,3*np.pi,0.1)
y = np.sin(x)

plt.plot(x, y)
plt.show()
```

```python
#  *** Plotting Multiple Lines: ***

# Compute the x and y coordinates for points on sine and cosine curves
# and plot them on the same graph using matplotlib.
# Add x and y labels to the graph as well.

x = np.arange(0, 3*np.pi, 0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)

plt.plot(x, y_sin)
plt.plot(x, y_cos)

plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.legend(["sine", "cosine"])

plt.show()
```

```python
#  *** Plotting Subplots: ***

# Compute the x and y coordinates for points on sin & cos curves.
# Set up a subplot grid that has height 2 and width 1,
# and set the first such subplot as active.
# Plot the sine and cosine graphs.

x = np.arange(0, 3*np.pi, 0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)

plt.subplot(2, 1, 1)
plt.plot(x, y_sin)
plt.title("sine")

plt.subplot(2, 1, 2)
plt.plot(x, y_cos)
plt.title("cosine")

plt.show()
```
